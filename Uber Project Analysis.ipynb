#Adding New Month Column

```python
df['Month']=pd.DatetimeIndex(df['START_DATE']).month

month_label ={1.0:'jan',2.0:'feb',3.0:'Mar',4.0:'Apr',5.0:'May',6.0:'Jun',7.0:'Jul',8.0:'Aug',9.0:'Sep',10.0:'Oct',11.0:'Nov',12.0:'Dec'}

df['Month']=df['Month'].map(month_label)

mon=df['Month'].value_counts(month_label)
```

```python
df.head()
```

```python
df.tail()
```

#In which months do people book uber rides less frequently?

```python
# Group the data by month and count the rides
monthly_rides = df.groupby('Month')['START_DATE'].count().reset_index()
monthly_rides.columns = ['Month', 'Ride_Count']  # Rename columns for clarity
monthly_rides = monthly_rides.sort_values(by=['Ride_Count'], ascending=True)
```

```python
fig = px.line(
    monthly_rides,
    x='Month',
    y='Ride_Count',
    title='Uber Rides Frequency by Month'
)

fig.show()
```

#Adding New Column Week.

```python
df['Day']=df.START_DATE.dt.weekday

data_label={0:'Mon',1:'Tue',2:'Wed',3:'Thu',4:'Fri',5:'Sat',6:'Sun'}

df['Day']=df['Day'].map(data_label)
```

```python
df.head()
```

#On which days of the week do people book uber rides the most?

```python
day_label=df['Day'].value_counts()
```

```python
fig=px.bar(day_label,x=day_label.index,y=day_label.values,title='Uber Rides Frequency by Day')
fig.show()
```
